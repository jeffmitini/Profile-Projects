# Linear Regression: Predicting car Prices

# Introduction

In this file, I create a linear regression model that predicts the price of a car, based on several variables like year of make, mileage, engine size and miles per gallon(mpg). 

The data is from Kaggle, titled ‘Ford Car Price Prediction’ (Quku, 2022). The dataset is a collection of Ford cars, their models, prices, mileage, and other characteristics. It has 9 fields and 17966 records. 

Below I import the libraries, the data from CSV, and 

# Importing Libraries and data

```python
#importing libraries

import numpy as np
import pandas as pd
import scipy.stats as stats
import math
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Import sample data
ford = pd.read_csv("ford.csv")
print(ford)
```

The description of the dataset reveals that there are 17966 records, making a very good sample. 

```python
print(ford.describe())
```

# Testing relation between price and other variables

The code that follows tests the relationship between our target variable, 'Price' to the rest of the variables using with scatter plot. The correlation coefficient was computed to see the strength of the relationship. 

```python
#scatter plot price vs engine year of make

plt.scatter(ford.year, ford.price)
plt.title("Relationship between Price and Year of Make")
plt.xlabel("Year of Make")
plt.ylabel("Price")
plt.show

#obtain m (slope) and b(intercept) of linear regression
m, b = np.polyfit(ford.year, ford.price, 1)

#add linear regression line to scatterplot 
plt.plot(ford.year, m*ford.year+b, color="k")

stats.linregress(ford.year, ford.price)
```

```python
#scatter plot price vs engine mileage

plt.scatter(ford.mileage, ford.price)
plt.title("Relationship between Price and Mileage")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show

plt.show

#obtain m (slope) and b(intercept) of linear regression
m, b = np.polyfit(ford.mileage, ford.price, 1)

#add linear regression line to scatterplot 
plt.plot(ford.mileage, m*ford.mileage+b, color="k")

stats.linregress(ford.mileage, ford.price)
```

```python
#scatter plot price vs tax

plt.scatter(ford.tax, ford.price)
plt.show

#obtain m (slope) and b(intercept) of linear regression
m, b = np.polyfit(ford.tax, ford.price, 1)

#add linear regression line to scatterplot 
plt.plot(ford.tax, m*ford.tax+b)

stats.linregress(ford.tax, ford.price)
```

```python
#scatter plot price vs mpg

plt.scatter(ford.mpg, ford.price)
plt.show

#obtain m (slope) and b(intercept) of linear regression
m, b = np.polyfit(ford.mpg, ford.price, 1)

#add linear regression line to scatterplot 
plt.plot(ford.mpg, m*ford.mpg+b)

stats.linregress(ford.mpg, ford.price)
```

```python
#scatter plot price vs engine size

plt.scatter(ford.engineSize, ford.price)
plt.show

#obtain m (slope) and b(intercept) of linear regression
m, b = np.polyfit(ford.engineSize, ford.price, 1)

#add linear regression line to scatterplot 
plt.plot(ford.engineSize, m*ford.engineSize+b)

stats.linregress(ford.engineSize, ford.price)
```

After testing the relationship of each variable to that of price, it was found that Year of Make and Mileage have the strongest relationships with price. It was noted however that all variables had a linear relationship with price. 

Year of make shows a Correlation Coefficient of 0.6360093404474128 and a Coefficient of Determination of 0.40450. That means there is a strong positive relationship between year of make and price, approximately 40% of the variation in price can be explained by the variation in year of make, using the best-fit regression line (Illowsky& Dean, 2022)

Mileage shows a Correlation Coefficient of 0.5306591212413305 and a Coefficient of Determination of 0.28159. That means there is a strong positive relationship between mileage and price, approximately 28% of the variation in price can be explained by the variation in mileage, using the best-fit regression line (Illowsky& Dean, 2022)

After trying regression models with different combination of variables, it was found that the best result was achieved by using the five variables of year, mileage, engine size, tax and mpg. Below is the regression models using the 5 variables.


# Creating Regression model with 5 variables

```python
#Create regression model with 5 variables

x = ford[['year','mileage', 'engineSize', 'tax', 'mpg']]
y = ford['price']
 
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(x.values, y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

```

# Testing the Model

The code below tests the model to predict a 2017 ford. The model goes on to predict the price for the vehicle to be $11,042. 

```python
#Predict price

predictedprice = regr.predict([[2017, 15944, 1, 150, 57.7]])

print(predictedprice)
```

Finally, to check the co-efficient of determination, also called R squared of the model, just like we did for the linear regression, we run the code as shown below:

```python
#calculate R-squared of regression model
r_squared = regr.score(x.values, y)

#view R-squared value
print(r_squared)
```

```python
#display adjusted R-squared
1 - (1-regr.score(x.values, y))*(len(y)-1)/(len(y)-x.shape[1]-1)
```

Notice that the code was run twice, first was the R-squared and the second one was adjusted R-squared. Notice that the result is almost the same. We can conclude that approximately 73% of the variation in car price can be explained by the variation in the 5 variables we have used in the model, that is, year of make, miles per gallon, engine size, mileage and tax (Illowsky& Dean, 2022)
